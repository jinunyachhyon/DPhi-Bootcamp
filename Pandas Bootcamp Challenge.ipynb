# Task 1

```python
###Task 1

import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/covid_19_data.csv") #Read the dataset and name is as df
```

```python
df.head(10)#Print first 10 observations of the dataset
```

```python
#Check for the data types of the all the variables in the dataset
df.dtypes
```

```python
#Convert columns ['ObservationDate'] and ['Last Update'] to from object to datetime data type.
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
df['Last Update'] = pd.to_datetime(df['Last Update'])
df.dtypes
```

# Task 2

```python
#Create a subset of the data df, only taking observations which was last updated on 2020-06-13 03:33:14 
#and name this dataframe as updated_data.

updated_data = df[df['Last Update'] == '2020-06-13 03:33:14']
```

```python
updated_data
```

```python
#Print the total no. of observations and unique country records in updated_data

print(updated_data.shape) #total observations

unique= updated_data['Country/Region'].unique() #unique country records
print(unique)
```

# Task 3

```python
"""Create a subset of data from the dataframe updated_data, 
only taking the top 10 countries which have the maximum number of confirmed cases with features -
'Country/Region', 'Confirmed', 'Deaths' and 'Recovered'. Name this dataframe as 'top_10'"""

updated_data=updated_data.sort_values(by='Confirmed', ascending = False)
updated_data.head()
```

```python
top_10=updated_data.iloc[0:10,[3,5,6,7]]
top_10
```

```python
#print the name of the countries in 'top_10' dataset
top_10['Country/Region']
```

```python
"""Which country has the lowest number of confirmed cases among the ten countries in top_10 dataframe? 
Hint: use groupby() to group all the observations with same country, 
sum all the observations and then sort the dataframe by confirmed cases."""

a=top_10.groupby('Country/Region').sum() #group all the observations and sum
```

```python
a.sort_values(by = 'Confirmed') #sort by confirmed cases
```

```python
a["Confirmed"].min()
```

```python
a[a["Confirmed"]==a["Confirmed"].min()]
```

# Task 4

```python
"""Add two columns in 'top_10' dataframe - 'Recovered_percentage' and 'Deaths_percentage'
where

'Recovered_percentage' = (Recovered cases / Confirmed cases) * 100

'Death_percentage' = (Deaths cases / Confirmed cases) * 100"""

top_10['Recovered_percentage']=(top_10['Recovered'] / top_10['Confirmed']) * 100
top_10['Deaths_percentage']=(top_10['Deaths'] / top_10['Confirmed']) * 100
```

```python
top_10
```

```python
#Print the top 3 countries with highest recovery precent

top_10_recovered=top_10.sort_values(by='Recovered_percentage',ascending = False)
top_10_recovered
```
